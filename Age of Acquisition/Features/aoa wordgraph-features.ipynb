```python
def avg_con_k(filename):
    
    graph = readinggraphs(filename)
    
    num_nodes = len(graph.nodes)
    num_edges = len(graph.edges)
    
    k = (2*num_edges)/num_nodes
    print(k)
    
    return k
```
